# Fit the spectrum

Define a model.

$$
y = A \cdot \text{exp} \Big\{ - \frac{(x-B)^2}{2C^2} \Big\}
$$

```python
def gaussian(x, a, b, c):
    """Gaussian function"""
    return a * np.exp(-((x - b)**2/(2*c**2)))
```

Fit the model

```python
p0 = [500, 10, 10]
params, cov = scipy.optimize.curve_fit(gaussian, wavelength, intensity, p0)
```

plot the model.

```python
xmodel = np.linspace(300,700,1000)
ymodel = gaussian(xmodel, *params)

plt.plot(xmodel, ymodel, "r")
plt.plot(wavelength, intensity)
plt.xlabel("wavelength")
plt.ylabel("intensity")
plt.title("spectrum")
```

Widget to get find an appropriate fit.

```python
@ipywidgets.interact
def fitter(file=glob.glob("*.xlsx"), a=(0,3000,100), b=(0,1000, 100), c=(0,1000,100)):
    # Load data
    df = pd.read_excel(file)
    i = int(file[-6]) - 1
    wavelength = df["xdata"]
    intensity = df["ydata"]
    
    # Fit model
    p0 = [a,b,c]
    params, cov = scipy.optimize.curve_fit(gaussian, wavelength, intensity, p0)
    xmodel = np.linspace(300,700,1000)
    ymodel = gaussian(xmodel, *params)

    plt.plot(wavelength, intensity, color="C{}".format(i), alpha=.7)

    plt.plot(xmodel, ymodel, "r", linewidth=4)
    plt.xlabel("wavelength")
    plt.ylabel("intensity")
    plt.title("spectrum")
    plt.show()
    
    out = pd.DataFrame(params, index=["a","b","c"], columns=["Parameters"])
    return out
```
